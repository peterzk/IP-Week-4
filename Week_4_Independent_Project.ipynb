## Univariate Analysis 

This univariate analysis is for Bluecars taken, Utilib taken and Utilib 14 taken

```
# Creating a new dataframe with columns that we want to use

auto_univ = auto_df[['Postal_code','Day_of_week','Month','Day_type','BlueCars_taken_sum','Utilib_taken_sum','Utilib_14_taken_sum']]

auto_univ.head()
```

## Hypothesis Testing and Analysis

```
#Copying the dataset that we are going to use for hypothesis testing

auto_hypothesis = auto_biv_out.copy()
```

```
#checking the shape of our dataset

auto_hypothesis.shape
```

* In the hypothesis testing, we are interested in the BlueCars.
* The goal is to test the claim that the average number of BlueCars taken during week days is different from the average number of BlueCars taken during the weekend.

* Null hypothesis: H 0: μ 1 = μ 2
* Alternative hypothesis: H a : μ 1 ≠ μ 2
* Significance level is 5%

```
# To generate two samples for our hypothesis testing, we need to separate
# the dataset based on the Day type

Auto_Weekday = auto_hypothesis.loc[auto_hypothesis['Day_type']==0] 

Auto_Weekend = auto_hypothesis.loc[auto_hypothesis['Day_type']==1]
```

The following formula shall be used to determine the sample size to use in our hypothesis testing.

> $n=  n_0/(1+ ((n_0-1)/N)$

> $n$ is the sample size

> $n_0$ is Cochran`s recommended sample size which is always 385

> $N$ is the population

> Using the formula above, with $N$ being 16085. The ideal sample size $n$ is 376.

```
# We use random sampling to get a ramdom sample from our datasets

Auto_Weekday_sample = Auto_Weekday.sample(n = 376, replace="False")

Auto_Weekend_sample = Auto_Weekend.sample(n = 376, replace="False")
```

```
# Checking the shape of our samples to see if they have the right sample size

print(Auto_Weekday_sample.shape)
print(Auto_Weekend_sample.shape)
```

```
print(Auto_Weekday_sample['BlueCars_taken_sum'].mean())
print(Auto_Weekend_sample['BlueCars_taken_sum'].mean())
```

```
# We import the modules that we are going to use to find our zscore and the p value
# We are going to use the the z test because we have a large sample with more than 30 values
from scipy import stats
from statsmodels.stats import weightstats as stests

ztest ,pval1 = stests.ztest(Auto_Weekday_sample['BlueCars_taken_sum'], Auto_Weekend_sample['BlueCars_taken_sum'], value=0,alternative='two-sided')
print(float(pval1))
print(float(ztest))

if pval1 < 0.05:
    print("Reject null hypothesis")
else:
    print("Do not reject null hypothesis")
```

From the results of our hypothesis test, we can conclude that the average number of BlueCars taken on weekdays is not the same as the average number of BlueCars taken during the weekends.
